```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import arff
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans, HDBSCAN
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 50)
```

# Data Loading

```python
ROOT = Path.cwd()

if ROOT.name == 'notebooks':
    DATA_PATH = ROOT.parent / 'credit.arff'
elif ROOT.name == 'statistical-data-analysis':
    DATA_PATH = 'credit.arff'
else:
    DATA_PATH = 'statistical-data-analysis/credit.arff'
    
with open(DATA_PATH, 'r') as f:
    arff_data = arff.load(f)
    
data = arff_data['data']
column_names = [attr[0] for attr in arff_data['attributes']]
    
df = pd.DataFrame(data, columns=column_names)
df.head()
```

# Feature Preprocessing

```python
df.info()
```

```python
# not sure about using - Income_Verified, Inquiries_Last_6Mo, Months_Since_Deliquency, Total_Accounts
feature_columns = ['Loan_ID', 'Loan_Amount_Requested', 'Length_Employed', 'Annual_Income', 'Purpose_Of_Loan', 'Debt_To_Income',
                   'Number_Open_Accounts', 'Gender']
X = df[feature_columns].copy()

numeric_features = X.select_dtypes(include='number').columns.tolist()
categorical_features = X.select_dtypes(exclude='number').columns.tolist()

# simple imputer chooses strategy for dealing with NaN values, standard scaler scales by removing the mean and scaling to unit variance
# one hot encoder encodes categorical features as a one-hot numeric array
preprocessor = ColumnTransformer([
    ('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), numeric_features),
    ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))]), categorical_features),
])

X_prepared = preprocessor.fit_transform(X)
X_prepared.shape
```

# Clustering Analysis

### KMeans clustering

```python
from sklearn.metrics import silhouette_score

inertias = []
silhouette_scores = []
k_range = range(2, 31)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    clusters = kmeans.fit_predict(X_prepared)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_prepared, labels=clusters))
    
k_final = k_range[np.argmax(silhouette_scores)]
kmeans_final = KMeans(n_clusters=k_final, random_state=42, n_init=20)
clusters = kmeans_final.fit_predict(X_prepared)

analysis_df = X.copy()
analysis_df['cluster'] = clusters
print(f"Selected k={k_final} (highest Silhouette Score), cluster sizes:")
print(analysis_df.groupby('cluster').size())
    
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    
ax = axs[0]
ax.plot(k_range, inertias, marker='o')
ax.set_xlabel('Number of clusters k')
ax.set_ylabel('Inertia')
ax.set_title('Elbow Method')
ax.set_xticks(k_range)
ax.grid(True)

ax = axs[1]
ax.plot(k_range, silhouette_scores, marker='o', color='orange')
ax.set_xlabel('Number of clusters k')
ax.set_ylabel('Silhouette Score')
ax.set_title('Silhouette Method')
ax.set_xticks(k_range)
ax.grid(True)
```

```python
analysis_df.groupby('cluster')[numeric_features].mean()
```

### HDBSCAN

```python
hdbscan_model = HDBSCAN(min_cluster_size=50)

clusters_hdbscan = hdbscan_model.fit_predict(X_prepared)
analysis_df_hdbscan = X.copy()
analysis_df_hdbscan['cluster'] = clusters_hdbscan

analysis_df_hdbscan.groupby('cluster')[numeric_features].mean()
```

### KMeans vs HDBSCAN comparison

```python
from sklearn.metrics import silhouette_score

mask = clusters_hdbscan != -1
n_noise = (clusters_hdbscan == -1).sum()
n_clusters_hdbscan = len(set(clusters_hdbscan[mask]))

s_kmeans = silhouette_score(X_prepared, clusters)
s_hdbscan = silhouette_score(X_prepared[mask], clusters_hdbscan[mask])

print(f"KMeans (k=2):                        Silhouette = {s_kmeans:.4f}")
print(f"HDBSCAN ({n_clusters_hdbscan} klastrów, -{n_noise} szum):  Silhouette = {s_hdbscan:.4f}")
```

# Data PCA

### 2 components PCA

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_prepared)

plot_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'cluster': clusters.astype(str),
})

sns.scatterplot(data=plot_df, x="PC1", y="PC2", hue="cluster")
plt.title("Clusters after PCA")
plt.show()
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_prepared)

plot_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'cluster': clusters_hdbscan.astype(str),
})

sns.scatterplot(data=plot_df, x="PC1", y="PC2", hue="cluster",
                palette="tab20", alpha=0.6, s=15, linewidth=0)
plt.title("Clusters (HDBSCAN) after PCA")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=7)
plt.tight_layout()
plt.show()
```

# Data UMAP and TSNE

```python
from sklearn.manifold import TSNE
import umap  

tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_jobs=-1, init='pca', learning_rate='auto')
X_tsne = tsne.fit_transform(X_prepared)

reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42, transform_seed=42)
X_umap = reducer.fit_transform(X_prepared)

income = analysis_df["Annual_Income"]
income_plot = income.fillna(income.median())
hue_lo, hue_hi = income_plot.min(), income_plot.quantile(0.9)

fig, axes = plt.subplots(2, 3, figsize=(20, 12))
titles_row = ['t-SNE', 'UMAP']
X_reds = [X_tsne, X_umap]

for row, (X_red, row_title) in enumerate(zip(X_reds, titles_row)):

    sns.scatterplot(x=X_red[:, 0], y=X_red[:, 1],
                    hue=income_plot, hue_norm=(hue_lo, hue_hi),
                    palette="viridis", alpha=0.6, linewidth=0,
                    legend=False, s=10, ax=axes[row, 0])
    axes[row, 0].set_title(f"{row_title} — Annual Income")

    sns.scatterplot(x=X_red[:, 0], y=X_red[:, 1],
                    hue=clusters.astype(str),
                    palette="tab10", alpha=0.6, linewidth=0,
                    s=10, ax=axes[row, 1], legend=True)
    axes[row, 1].set_title(f"{row_title} — KMeans (k={k_final})")
    axes[row, 1].legend(title="Klaster", fontsize=7, title_fontsize=8)

    sns.scatterplot(x=X_red[:, 0], y=X_red[:, 1],
                    hue=clusters_hdbscan.astype(str),
                    palette="tab20", alpha=0.6, linewidth=0,
                    s=10, ax=axes[row, 2], legend=False)
    axes[row, 2].set_title(f"{row_title} — HDBSCAN ({len(set(clusters_hdbscan[clusters_hdbscan != -1]))} klastrów)")

norm = plt.Normalize(hue_lo, hue_hi)
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=[axes[0, 0], axes[1, 0]], label="Annual_Income", shrink=0.6)

plt.tight_layout()
plt.show()
```
